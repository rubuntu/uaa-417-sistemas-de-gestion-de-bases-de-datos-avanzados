**Ejercicios básicos e intermedios de matplotlib con los datos de muestra de Google Colab**

---

**Ejercicio 1:**

Cargar el archivo `california_housing_train.csv` y crear un gráfico de dispersión (scatter plot) de `median_income` vs. `median_house_value`.

---

**Ejercicio 2:**

Usando el mismo conjunto de datos, crear un histograma de la columna `median_house_value` con 50 bins y agregarle títulos y etiquetas a los ejes.

---

**Ejercicio 3:**

Generar un gráfico de barras que muestre el número de casas (`households`) en cada rango de edad (`housing_median_age`), agrupando las edades en bins de 5 años.

---

**Ejercicio 4:**

Crear un mapa de calor (heatmap) que muestre la correlación entre las variables numéricas del DataFrame `california_housing_train.csv`.

---

**Ejercicio 5:**

Usar el archivo `mnist_train_small.csv` para mostrar la imagen del dígito en la primera fila, convirtiendo los valores de los píxeles en una matriz de 28x28 y visualizándola con `imshow`.

---

**Ejercicio 6:**

Crear subplots que muestren imágenes de los dígitos del 0 al 9, seleccionando una muestra de cada uno del conjunto de datos `mnist_train_small.csv`.

---

**Ejercicio 7:**

Cargar los datos de `anscombe.json` y crear gráficos de dispersión para cada uno de los cuatro conjuntos de datos de Anscombe, incluyendo la línea de regresión en cada gráfico.

---

**Ejercicio 8:**

Usando `california_housing_train.csv`, crear un boxplot de `median_house_value` para diferentes categorías de `ocean_proximity` (nota: si esta columna no existe, crear categorías basadas en la ubicación geográfica).

---

**Ejercicio 9:**

Crear una gráfica de líneas que muestre la tendencia de `median_house_value` en función de la `latitude`, ordenando los datos por `latitude`.

---

**Ejercicio 10:**

Generar un gráfico de pastel (pie chart) que muestre la proporción de viviendas en diferentes rangos de `median_income`, por ejemplo: bajos ingresos (<2), ingresos medios (2-5), ingresos altos (>5).

---

Estos ejercicios te ayudarán a practicar la visualización de datos utilizando matplotlib y a interpretar diferentes tipos de gráficos basados en conjuntos de datos reales disponibles en Google Colab.
